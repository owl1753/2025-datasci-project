```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 읽기
df = pd.read_csv('data2.csv')

# 환자시군구명과 구급활동구분명에 따른 데이터 분석
emergency_by_region = df.groupby(['patnt_sigungu_nm', 'relif_act_se_nm']).size().unstack(fill_value=0)

# 시각화 1: 스택 바 차트
plt.figure(figsize=(15, 8))
emergency_by_region.plot(kind='bar', stacked=True)
plt.title('지역별 응급 유형 분석')
plt.xlabel('환자 시군구')
plt.ylabel('발생 건수')
plt.xticks(rotation=45)
plt.legend(title='구급활동구분명', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 시각화 2: 히트맵
plt.figure(figsize=(15, 8))
sns.heatmap(emergency_by_region, annot=True, fmt='d', cmap='YlOrRd')
plt.title('지역별 응급 유형 히트맵')
plt.xlabel('구급활동구분명')
plt.ylabel('환자 시군구')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 상세 통계 출력
print("\n지역별 응급 유형 통계:")
print(emergency_by_region)

# 전체 응급 유형 비율
print("\n전체 응급 유형 비율:")
total_emergency = df['relif_act_se_nm'].value_counts()
print(total_emergency)

# 지역별 총 응급 건수
print("\n지역별 총 응급 건수:")
region_total = df['patnt_sigungu_nm'].value_counts()
print(region_total)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 읽기
df = pd.read_csv('data2.csv')

# 결측치 제거
df_clean = df.dropna(subset=['patnt_occp_nm', 'acdnt_ocrn_place'])

# 직업군별 사고 발생 장소 분석
accident_by_occupation = pd.crosstab(df_clean['patnt_occp_nm'], df_clean['acdnt_ocrn_place'])

# 시각화 1: 히트맵
plt.figure(figsize=(15, 10))
sns.heatmap(accident_by_occupation, annot=True, fmt='d', cmap='YlOrRd')
plt.title('직업군별 사고 발생 장소 분석')
plt.xlabel('사고 발생 장소')
plt.ylabel('직업군')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화 2: 스택 바 차트
plt.figure(figsize=(15, 8))
accident_by_occupation.plot(kind='bar', stacked=True)
plt.title('직업군별 사고 발생 장소 분포')
plt.xlabel('직업군')
plt.ylabel('발생 건수')
plt.legend(title='사고 발생 장소', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 상세 통계 출력
print("\n직업군별 사고 발생 장소 통계:")
print(accident_by_occupation)

# 직업군별 총 사고 건수
print("\n직업군별 총 사고 건수:")
print(accident_by_occupation.sum(axis=1))

# 사고 발생 장소별 총 건수
print("\n사고 발생 장소별 총 건수:")
print(accident_by_occupation.sum(axis=0))

# 각 직업군별 가장 빈번한 사고 발생 장소
print("\n직업군별 가장 빈번한 사고 발생 장소:")
for occupation in accident_by_occupation.index:
    most_common = accident_by_occupation.loc[occupation].idxmax()
    count = accident_by_occupation.loc[occupation, most_common]
    print(f"{occupation}: {most_common} ({count}건)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 읽기
df = pd.read_csv('data2.csv')

# 상업시설 사고 데이터 필터링
commercial_accidents = df[df['acdnt_ocrn_place'] == '상업시설']

# 지역별 상업시설 사고 건수 분석
region_accidents = commercial_accidents['patnt_sigungu_nm'].value_counts()

# 시각화 1: 막대 그래프
plt.figure(figsize=(15, 8))
region_accidents.plot(kind='bar')
plt.title('지역별 상업시설 사고 발생 건수')
plt.xlabel('지역')
plt.ylabel('사고 발생 건수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시각화 2: 파이 차트 (상위 10개 지역)
plt.figure(figsize=(12, 8))
top_10_regions = region_accidents.head(10)
plt.pie(top_10_regions, labels=top_10_regions.index, autopct='%1.1f%%')
plt.title('상위 10개 지역의 상업시설 사고 비율')
plt.axis('equal')
plt.show()

# 상세 통계 출력
print("\n지역별 상업시설 사고 발생 건수:")
print(region_accidents)

# 전체 대비 비율 계산
total_commercial = len(commercial_accidents)
print("\n전체 상업시설 사고 건수:", total_commercial)
print("\n지역별 상업시설 사고 비율:")
print((region_accidents / total_commercial * 100).round(2), "%")

# 상위 5개 지역의 상세 정보
print("\n상위 5개 지역의 상세 정보:")
top_5_regions = region_accidents.head()
for region in top_5_regions.index:
    region_data = commercial_accidents[commercial_accidents['patnt_sigungu_nm'] == region]
    print(f"\n{region} 지역 상업시설 사고 정보:")
    print(f"총 사고 건수: {len(region_data)}건")
    print("주요 응급 유형:")
    print(region_data['relif_act_se_nm'].value_counts().head())
```
