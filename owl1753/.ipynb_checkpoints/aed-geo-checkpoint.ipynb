```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
```

```python
# 공간 데이터 파일 읽기  
gdf = gpd.read_file("emd.shp", encoding='cp949').to_crs('EPSG:4166')
```

```python
aed_data = pd.read_csv("../data/AED.csv")
```

```python
aed_geo = gpd.GeoDataFrame(
    aed_data, geometry=gpd.points_from_xy(aed_data['경도'], aed_data['위도']), crs="EPSG:4166"
)
```
