```python
import pandas as pd
import geopandas as gpd
```

```python
# 공간 데이터 파일 읽기  
gdf = gpd.read_file("./emd/emd.shp", encoding='cp949').to_crs('EPSG:4166')
```

```python
gdf
```

```python
pop_gdf = gpd.read_file("./population/population.shp", encoding='cp949').set_crs('EPSG:5181').to_crs('EPSG:4166')
```

```python
pop_gdf
```

```python
pop_gdf.convex_hull.plot(color='gray', edgecolor="w")
```

```python
df = pd.DataFrame(gpd.sjoin(pop_gdf, gdf))
```

```python
df
```

```python
df.to_csv('population_with_geo.csv', index=False)
```
