```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
```

```python
# 공간 데이터 파일 읽기  
gdf = gpd.read_file("./emd/emd.shp", encoding='cp949').to_crs('EPSG:4166')
```

```python
gdf
```

```python
emergency_data = pd.read_csv("./중증응급질환 구급출동 현황.csv")
```

```python
emergency_data = emergency_data[emergency_data['hrtarst_nm'] == '심정지']
```

```python
emergency_data
```

```python
gdf[['EMD_KOR_NM', 'geometry']]
```

```python
emergency_geo = gpd.GeoDataFrame(
    emergency_data, geometry=gpd.points_from_xy(emergency_data['gis_x_axis'], emergency_data['gis_y_axis']), crs="EPSG:5186"
).to_crs('EPSG:4166')
```

```python
emergency_geo[['sigungu_nm', 'geometry']]
```

```python
df = pd.DataFrame(gpd.sjoin(emergency_geo, gdf))
```

```python
df
```

```python
df.to_csv('./emergency_with_geo.csv', index=False)
```
